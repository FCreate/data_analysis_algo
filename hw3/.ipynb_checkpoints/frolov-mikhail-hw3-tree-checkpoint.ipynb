```python
from time import time

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from scipy import optimize
from sklearn.metrics import accuracy_score
from sklearn.model_selection import KFold
from sklearn.tree import DecisionTreeClassifier

%matplotlib inline
```

```python
class MyDecisionTreeClassifier:
    NON_LEAF_TYPE = 0
    LEAF_TYPE = 1

    def __init__(self, min_samples_split=2, max_depth=None, sufficient_share=1.0, criterion='gini', max_features=None):
        self.tree = dict()
        self.min_samples_split = min_samples_split
        self.max_depth = max_depth
        self.sufficient_share = sufficient_share
        self.num_class = -1
        if criterion == 'gini':
            self.G_function = self.__gini
        elif criterion == 'entropy':
            self.G_function = self.__entropy
        elif criterion == 'misclass':
            self.G_function = self.__misclass
        else:
            print 'invalid criterion name'
            raise

        if max_features == 'sqrt':
            self.get_feature_ids = self.__get_feature_ids_sqrt
        elif max_features == 'log2':
            self.get_feature_ids = self.__get_feature_ids_log2
        elif max_features == None:
            self.get_feature_ids = self.__get_feature_ids_N
        else:
            print 'invalid max_features name'
            raise

    def __gini(self, l_c, l_s, r_c, r_s):
        l_s = l_s.astype('float')
        r_s = r_s.astype('float')
        return 1 - np.sum( l_c**2 / l_s + r_c**2 / r_s, axis=1)
    
    def __entropy(self, l_c, l_s, r_c, r_s):
        l_s = l_s.astype('float')
        r_s = r_s.astype('float')
        return -(np.sum(l_c * np.log((l_c + 1e-4) / l_s) + r_c * np.log((r_c + 1e-4) / r_s), axis=1)).reshape(-1, 1) / (l_s + r_s)

    def __misclass(self, l_c, l_s, r_c, r_s):
        return 1 - np.max(l_c / (l_s + r_s), axis=1) - np.max(r_c / (l_s + r_s), axis=1)

    def __get_feature_ids_sqrt(self, n_feature):
        feature_ids = range(n_feature)
        np.random.shuffle(feature_ids)
        return feature_ids[:int(sqrt(n_feature))]
        
    def __get_feature_ids_log2(self, n_feature):
        feature_ids = range(n_feature)
        np.random.shuffle(feature_ids)
        return feature_ids[:int(log2(n_feature))]

    def __get_feature_ids_N(self, n_feature):
        return range(n_feature)
    
    def __sort_samples(self, x, y):
        sorted_idx = x.argsort()
        return x[sorted_idx], y[sorted_idx]

    def __div_samples(self, x, y, feature_id, threshold):
        left_mask = x[:, feature_id] > threshold
        right_mask = ~left_mask
        return x[left_mask], x[right_mask], y[left_mask], y[right_mask]

    def __find_threshold(self, x, y):
        #Сортируем по X
        sorted_x, sorted_y = self.__sort_samples(x, y)
        #Количество кластеров
        class_number = np.unique(y).shape[0]
        
        #Отбрасываем слева и справа по min_samples_split
        splitted_sorted_y = sorted_y[self.min_samples_split:-self.min_samples_split]
        #Ищем индексы, на которых происходит смена класса
        r_border_ids = np.where(splitted_sorted_y[:-1] != splitted_sorted_y[1:])[0] + (self.min_samples_split + 1)
        #Массив имеет одинаковый класс, либо же он меньше, чем splitted_sorted_y,
        #то возвращаем impurity = +inf
        if len(r_border_ids) == 0:
            return float('+inf'), None
        
        #Находим количество элементов в одном блоке(одинаковые последовательно идущие элементы),
        #кроме последнего блока
        eq_el_count = r_border_ids - np.append([self.min_samples_split], r_border_ids[:-1])
        #количество правых границ массивов нулей, в каждом массиве class_number нулей
        one_hot_code = np.zeros((r_border_ids.shape[0], class_number))
        #Для всех блоков, кроме последнего ставим 1 в [i,j], где i-й блок имеет номер j
        one_hot_code[np.arange(r_border_ids.shape[0]), sorted_y[r_border_ids - 1]] = 1
        #Пересчитываем размеры блоков
        class_increments = one_hot_code * eq_el_count.reshape(-1, 1)
        #Возвращаем то, что отбросили из splitted_sorted
        class_increments[0] = class_increments[0] + np.bincount(sorted_y[:self.min_samples_split], minlength=class_number)
        
        #Вычисляем сколько элементов останется пойдет налево, а сколько направо
        l_class_count = np.cumsum(class_increments, axis=0)        
        r_class_count = np.bincount(y) - l_class_count
        l_sizes = r_border_ids.reshape(l_class_count.shape[0], 1)
        r_sizes = sorted_y.shape[0] - l_sizes

        #Ищем лучшее разбиение(наименьшее impurity) 
        gs = self.G_function(l_class_count, l_sizes, r_class_count, r_sizes)
        idx = np.argmin(gs)
    
        #Номер элемента правого класса
        left_el_id = l_sizes[idx][0]
        #Вернуть impurity и threshhold
        return gs[idx], (sorted_x[left_el_id-1] + sorted_x[left_el_id]) / 2.0
    def __leaf(self, y, node_id):
        max_probs = np.bincount(y).argmax()
        probs = np.bincount(y)/y.shape
        self.tree[node_id] = self.LEAF_TYPE, max_probs, probs
        pass
    
    def __fit_node(self, x, y, node_id, depth, pred_f=-1):
        # Ваш код
        # Необходимо использовать следующее:
        # self.LEAF_TYPE
        # self.NON_LEAF_TYPE

        # self.tree
        # self.max_depth
        # self.sufficient_share
        # self.min_samples_split

        # self.get_feature_ids
        # self.__find_threshold
        # self.__div_samples
        # self.__fit_node
        if (depth ==self.max_depth):
            self.__leaf(y, node_id)  
            return
        if (x.shape[0] < 2 * self.min_samples_split + 2):
            self.__leaf(y, node_id)
            return
        if ((np.bincount(y).max()/y.shape) > self.sufficient_share):
            self.__leaf(y, node_id)
            return
        
        feature_ids = self.get_feature_ids(x.shape[1])
        thresholds = np.array([self.__find_threshold(x[:, i], y) for i in feature_ids])
        best_split = thresholds[:,0].argmin()
        impurity, threshold = thresholds[best_split]
        left, right, left_classes, right_classes = self.__div_samples(x, y, feature_ids[best_split], threshold)
        if (left.shape[0] < self.min_samples_split):
            self.__leaf(y, node_id)
            return
        if (right.shape[0] < self.min_samples_split):
            self.__leaf(y, node_id)
            return
        
        self.tree[node_id] = self.NON_LEAF_TYPE, feature_ids[best_split], threshold
        self.__fit_node(left, left_classes, 2 * node_id + 1, depth + 1, pred_f)
        self.__fit_node(right, right_classes, 2 * node_id + 2, depth + 1, pred_f)              
    
    def fit(self, x, y):
        self.num_class = np.unique(y).size
        self.__fit_node(x, y, 0, 0) 

    def __predict_class(self, x, node_id):
        node = self.tree[node_id]
        if node[0] == self.__class__.NON_LEAF_TYPE:
            _, feature_id, threshold = node
            if x[feature_id] > threshold:
                return self.__predict_class(x, 2 * node_id + 1)
            else:
                return self.__predict_class(x, 2 * node_id + 2)
        else:
            return node[1]

    def __predict_probs(self, x, node_id):
        node = self.tree[node_id]
        if node[0] == self.__class__.NON_LEAF_TYPE:
            _, feature_id, threshold = node
            if x[feature_id] > threshold:
                return self.__predict_probs(x, 2 * node_id + 1)
            else:
                return self.__predict_probs(x, 2 * node_id + 2)
        else:
            return node[2]
        
    def predict(self, X):
        return np.array([self.__predict_class(x, 0) for x in X])
    
    def predict_probs(self, X):
        return np.array([self.__predict_probs(x, 0) for x in X])

    def fit_predict(self, x_train, y_train, predicted_x):
        self.fit(x_train, y_train)
        return self.predict(predicted_x)
```

```python
df = pd.read_csv('./cs-training.csv', sep=',').dropna()
df.head()
```

```python
x = df.as_matrix(columns=df.columns[1:])
y = df.as_matrix(columns=df.columns[:1])
y = y.reshape(y.shape[0])
```

```python
my_clf = MyDecisionTreeClassifier(min_samples_split=2)
clf = DecisionTreeClassifier(min_samples_split=2)
```

## Проверка скорости работы

```python
t1 = time()
my_clf.fit(x, y)
t2 = time()
print(t2 - t1)

t1 = time()
clf.fit(x, y)
t2 = time()
print(t2 - t1)
```

## Проверка качества работы

```python
gkf = KFold(n_splits=5, shuffle=True)
```

```python
for train, test in gkf.split(x, y):
    X_train, y_train = x[train], y[train]
    X_test, y_test = x[test], y[test]
    my_clf.fit(X_train, y_train)
    print(accuracy_score(y_pred=my_clf.predict(X_test), y_true=y_test))
```

```python
for train, test in gkf.split(x, y):
    X_train, y_train = x[train], y[train]
    X_test, y_test = x[test], y[test]
    clf.fit(X_train, y_train)
    print(accuracy_score(y_pred=clf.predict(X_test), y_true=y_test))
```
